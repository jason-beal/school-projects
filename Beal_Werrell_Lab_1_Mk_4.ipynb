## Implementation

```
data = ["V906 CAR",2458235.023,6.873,2458251.989,8.025,2458272.936,9.115,
        2458508.233,10.577,"V392 PER",2458239.307,7.27, 2458242.681,8.6708,
        2458251.359,9.777,2458270.37,11.807,"V357 Mus",2458133.657,6.779,
        2458146.987,7.92,2458165.932,9.81,2458175.964,12.019]

## This code takes an input data list of multiple data types (strings and floats)
## in the item order [name], [date], [intensity], [date], [intensity], etc. The 
## output is a dictionary of nova name's with nested date:intensity pairs.

## Initialization
i_range = range(1,len(data))
nova_dict = {}
nova_data = {}
nova_name = data[0]
date_read = False

## Data manipulation
for i in i_range:
  if type(data[i]) == str:
    nova_dict[nova_name] = nova_data ## stores nova_data nested in nova_dict
    nova_name = data[i] ## sets read string as nova_name
    nova_data = {}
    date_read = False
  else:
    if not date_read:
      nova_data[data[i]] = data[i+1] ## pairs observation date with intensity value
      date_read = True
    else:
      date_read = False
      if i == len(data)-1:
        nova_dict[nova_name] = nova_data ## this statement saves the data for the final nova

for key in nova_dict.keys():
    print(key, nova_dict[key])
```
